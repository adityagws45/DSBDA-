```python
# . df.sample()
# The .sample() function is used to randomly select rows from the DataFrame.

# You can specify how many rows to pick using:

# n â†’ exact number of rows

# frac â†’ fraction of the total number of rows

# 2. frac=0.1
# This means: take 10% of the total rows.

# For example, if your DataFrame has 500 rows, this will return 0.1 Ã— 500 = 50 rows.

# 3. random_state=1
# This is a seed for random number generation.

# Setting a fixed value (like 1) ensures reproducibility â€” you'll get the same random sample every time you run the code.

# If you remove this or use a different value, you'll get different rows each time.

# 4. subset_5 =
# Stores the sampled rows into a new DataFrame called subset_5

subset_5=df.sample(frac=0.1,random_state=1)
```

```python
# . .loc[0:4]
# .loc[] is a label-based indexing method. Here, it is used to select rows based on their index.

# 0:4 means select rows from index 0 to index 4 inclusive. This means:

submerge_1=df[['Total Interactions','share']].loc[0:4]
```

```python
submerge_1
```

```python
submerge_2=df[['share','like']].loc[3:7]
```

```python
submerge_2
```

```python
# pd.merge() is a function in Pandas used to combine two DataFrames based on a common column or index.
# submerge_1 and submerge_2 are two separate DataFrames (which you created earlier) that you want to combine (merge).
# submerge_1 contains the columns 'share' and 'like'.
# submerge_2 contains the columns 'share' and 'Total Interactions'.
# on='share':
# This specifies that the merge should be done on the 'share' column. Both submerge_1 and submerge_2 should have a column named 'share', and this column will be used to match rows between the two DataFrames.
# how='inner':
# how='inner' specifies that the inner join should be used for the merge.

# In an inner join, only the rows that have matching values in both DataFrames for the 'share' column will be included in the resulting DataFrame.
# The resulting DataFrame (merge_inner) will contain only the rows where the value of 'share' is common in both submerge_1 and submerge_2.
merge_inner=pd.merge(submerge_1,submerge_2,on='share',how='inner')
```

```python

merge_inner
```

```python
# Merging data using outer join
# . Outer Join (how='outer'):
# An outer join returns all rows from both DataFrames (submerge_1 and submerge_2), along with the matching rows where the 'share' column values are the same.

# If a row from one DataFrame does not have a corresponding match in the other DataFrame, the non-matching columns will have NaN values in the resulting DataFrame.

# 2. on='share':
# The merge operation is performed based on the 'share' column. It matches rows from both DataFrames where 'share' values are equal.
merge_outer=pd.merge(submerge_1,submerge_2,on='share',how='outer')
```

```python
merge_outer
```

```python
# The left join keeps all the rows from the left DataFrame (submerge_1) and tries to match them with rows from the right DataFrame (submerge_2) based on the share column.

# If thereâ€™s no match in the right DataFrame, the result will contain NaN for the columns from the right DataFrame
merge_left=pd.merge(submerge_1,submerge_2,on='share',how='left')
```

```python

merge_left
```

```python
# The right join keeps all the rows from the right DataFrame (submerge_2) and tries to match them with rows from the left DataFrame (submerge_1) based on the share column.

# If thereâ€™s no match in the left DataFrame, the result will contain NaN for the columns from the left DataFrame.

merge_right=pd.merge(submerge_1,submerge_2,on='share',how='right')
```
